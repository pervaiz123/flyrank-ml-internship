## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
print("""
# My Lane as an ML Task

Lane: Refresh / Content Opportunity Scoring

Task Type:
Ranking / Scoring

The goal is to rank pages according to their likelihood of needing review for refresh or optimization.

Instead of predicting a simple yes/no outcome, the system produces a priority score so content teams can review the most important pages first.
""")
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
print("""
# Target or Proxy

Target / Proxy:

A page is considered a review candidate when its trend direction is declining.

Proxy Target:
trend_direction == "down"

This proxy helps identify pages that may benefit from content review or refresh.
""")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
print("""
# Success Metric

Primary Metric:
Precision@50

Reason:
Content teams usually review only a limited number of pages. Precision@50 measures how many of the top-ranked pages are truly declining.

Higher Precision@50 means more useful recommendations and less wasted effort.
""")
```
